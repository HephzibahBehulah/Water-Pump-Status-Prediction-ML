### `Task Description`
Can you predict which water pumps are faulty to promote access to clean, potable water across Tanzania?
This is an intermediate-level practice competition.
---

- Build a model to predict whether a water pump is faulty.
- Handle missing values and encode categorical features.
- Train a classification model and evaluate performance on imbalanced data
- Focus on precision, recall, and F1-score
- Dataset : [Pump it Up: Data Mining the Water Table](https://www.drivendata.org/competitions/7/pump-it-up-data-mining-the-water-table/data/).
- GitHub Colab: https://github.com/HephzibahBehulah/Water-Pump-Prediction

---
### `Approach`
We'll build 6 models which are Logistic Regression, Decision Tree, Random forest, XGBoost, LightGBM and CastBoost for more precision

### `The labels in this dataset`

The labels in this dataset are simple. There are three possible values:
- functional - the waterpoint is operational and there are no repairs needed
- functional needs repair - the waterpoint is operational, but needs repairs
- non functional - the waterpoint is not operational

```python
# ---- 0. Columns with data types ----
col_types = {}
for col in df.columns:
    types = df[col].apply(type).value_counts()
    col_types[col] = {str(t): n for t, n in types.items()}
col_types
```

```python
# ---- 1. Missing value check before cleaning data ----
print(f'Missing value = {df.isnull().sum()}') # => missing value

print("Train missing:\n", df.isnull().sum().sum())
print("Test missing:\n", test_df.isnull().sum().sum())
```

```python
# ---- 2 Identify columns ----
numeric_cols = ['amount_tsh','gps_height','longitude','latitude','num_private',
                'population','construction_year']

categorical_cols = ['funder','installer','wpt_name','subvillage',
                    'public_meeting','scheme_management','scheme_name','permit']
```

```python
# --- 3. Identify target and features ---
target_col = 'status_group'
X = df.drop(columns=[target_col, 'id'])
y = df[target_col]
X_test = test_df.drop(columns=['id'])

# --- 4. Identify categorical columns ---
cat_cols = X.select_dtypes(include=['object']).columns
```

```python
# --- 5. Fix mixed types in categorical columns ---
for col in cat_cols:
    X[col] = X[col].astype(str).fillna('missing')
    X_test[col] = X_test[col].astype(str).fillna('missing')

# --- 6. Fill missing numeric values with 0 ---
num_cols = X.select_dtypes(include=['int', 'float']).columns
X[num_cols] = X[num_cols].fillna(0)
X_test[num_cols] = X_test[num_cols].fillna(0)
```

```python
# --- 7. Encode categorical variables ---
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X[cat_cols] = encoder.fit_transform(X[cat_cols])
X_test[cat_cols] = encoder.transform(X_test[cat_cols])
```

```python
# --- 8. Lock in cleaned data ---
def clean_data(data):
    # Fill categorical
    cat_cols = data.select_dtypes(include='object').columns
    for col in cat_cols:
        data[col] = data[col].fillna('missing')
    
    # Fill numerical
    num_cols = data.select_dtypes(include=['int64','float64']).columns
    for col in num_cols:
        data[col] = data[col].fillna(0)
    
    return data
```

```python
# --- 9. Clean Data-- 
df = clean_data(df)
test_df = clean_data(test_df)
```
